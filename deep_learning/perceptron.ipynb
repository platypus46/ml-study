### 생물신경망과 인공신경망  
**인간의 뉴런(신경계를 이루는 기본적인 단위 세포)**

위의 개념을 추상화하여 컴퓨터로 구현한 것이 퍼셉트론이다.

*깊은 인공신경망은 이러한 퍼셉트론의 병렬배치를 순차적 구조로 결합한 형태를 띄고 있다.*

### 퍼셉트론에 대해 이해하기위해 먼저 뉴런의 동작방식을 한번 살펴보자(결론만 봐도 상관없음)###  
뉴런이 휴지상태일때 일정 강도 이상의 신호가 발생하면 신경세포체에서 축삭으로 넘어가는 곳에서 나트륨 채널이 열리며 이온들이 몰려와 급격한 전위차가 생기고 이때 이온이 교환되어 활동 전위가 생긴다. 즉 전기 신호가 발생하고, 이런 식으로 말단까지 전위가 전달되면 말단 안의 작은 알 같은 소포체들이 끝쪽으로 붙어 터지면서 신경전달물질이 시냅스 틈으로 방출된다. 시냅스 후 뉴런의 가지 돌기에서 이 전달물질을 받아들이면 다음 세포로 신호가 전달된다.

*※물론 뉴런이 자극을 받아들이는 정도는 측정하기 어렵고 뉴런의 종류마다 다르며 확률적으로 신호가 전달되지 않기도 하는데 이는 일단 논외로 하자.*  
*(확률적신경망이라는 개념이 있긴함)* 

**결론: 화학적 신호, 전기적 신호와 같이 어떠한 형태로 신호가 전달되는가에 대한 차이와 신호 전달방식에서의 기계적 차이만 존재할뿐 큰 관점에서는 일정신호(임계값)을 넘기게되면 다음 단계로 신호를 전달한다는 맥락에서 퍼셉트론과 뉴런은 동일하다고 할 수 있다.**

### 퍼셉트론의 구조###  

- 입력  
1.i번째 노드는 특징 벡터 $x=(x_1,x_2,...,x_d)^T$의 요소 $x_i$를 담당한다.  
2.항상 1이 입력되는 편향(bias)노드를 포함한다.  
<br>
- 연산  
1.i번째 입력 노드와 출력노드를 연결하는 변은 가중치$w_i$를 가진다.  
2.퍼셉트론은 단일 층 구조라고 간주한다.  
<br>  
- 출력  
1.한 개의 노드에 의해 수치(+1 혹은 -1)을 출력한다.

### 퍼셉트론 수식을 통해서 이해하기###    
선형회귀는 뉴럴네트워크로 표현이 가능하며 아래의 이미지와 식을 보고 항상 1이 입력되는 편향노드가 왜 필요한지도 이해할 수 있다.  
<img src="image/퍼셉트론이해.png" width="300" height="300">  


행렬로 표현하면 $\begin{bmatrix}x_1&1\\x_2&1\\x_3&1 \end{bmatrix}\begin{bmatrix}a\\b\\ \end{bmatrix}=\begin{bmatrix}y_1\\y_2\\y_3 \end{bmatrix}$가 된다.  
위의 이미지에서는 a,b만 다루고 이를 찾는 것이 목적이지만 앞으로 a,b에 대해서만 다루는 것뿐만 아니라 파라미터가 여러개 일 수 있기때문에 이를 통틀어 가중치(weight)로 표현하여 식을 세우면 아래와 같이 세울 수 있다.  
### 수식 ###  
$s=w_0+\displaystyle\sum_{i=1}^{d}{w_ix_i}$  
그리고 이에 최종적으로 출력(활성 함수$\tau$)을 추가하게되면 퍼셉트론의 전체 동작과정을 나타낸 것이 된다.  
활성화 함수에는 여러가지가 있는데 그 중 계단함수(step function)를 사용하면 다음과 같이 표현이 가능하다.  
$y=\tau(s)$  
$\tau(s)=((s>=0) → 1,(s<0) → -1)$


### 예시를 통해 퍼셉트론의 동작원리를 알아보자.(계단함수를 활용)

$X_1=\begin{pmatrix}0\\0\\ \end{pmatrix},y_1=-1,X_2=\begin{pmatrix}1\\0\\ \end{pmatrix},y_2=1,X_3=\begin{pmatrix}0\\1\\ \end{pmatrix},y_3=1,X_4=\begin{pmatrix}1\\1\\ \end{pmatrix},y_4=1$  
<img src="image/퍼셉트론의동작.png" width="600" height="300">  
샘플4개를 하나씩 확인하면서 제대로 분류하는지 확인해보자.  
- $X_1: s=-0.5+0*1.0+0*1.0=-0.5, \tau(-0.5)=-1$
- $X_2: s=-0.5+1*1.0+0*1.0=0.5, \tau(0.5)=1$
- $X_3: s=-0.5+0*1.0+0*1.1=0.5, \tau(0.5)=1$
- $X_4: s=-0.5+1*1.0+0*1.1=1, \tau(1)=1$  
이 퍼셉트론은 OR게이트 문제에 대해 훈련집합을 100%성능으로 분류한다고 말할 수 있다.

## 활성화 함수  
계단함수는 극단적으로 반환값이 0과 1이고 중간 계층에서 값을 살리기 너무 힘들다.  
따라서 계단함수는 활성화함수로 잘 사용되지는 않는다.  
아래의 코드를 통해 계단함수를 제외한 몇몇의 활성화함수들을 알아보자.(sigmoid,tanh,relu)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return list(map(lambda x:math.tanh(x),x))

def relu(x):
    result=[]
    for ele in x:
        if(ele<=0):
            result.append(0)
        else:
            result.append(ele)
            
    return result

x=np.linspace(-4,4,100)

plt.subplot(1, 3, 1) 
plt.plot(x,sigmoid(x))

plt.subplot(1, 3, 2) 
plt.plot(x,tanh(x))

plt.subplot(1, 3, 3) 
plt.plot(x,relu(x))

plt.tight_layout()
plt.show
```

### sigmoid함수를 이용하여 OR게이트 구현해보기

```python
import tensorflow as tf
tf.random.set_seed(42)

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse

#데이터 준비
x=np.array([[0,0],[1,0],[0,1],[1,1]])
y=np.array([[0],[1],[1],[1]])

# 단층 퍼셉트론 구성
model=Sequential()
model.add(Dense(1,input_shape=(2,),activation='sigmoid'))

#모델 준비
model.compile(optimizer=SGD(),
             loss=mse,
             metrics=['acc'])  

#학습시키기
model.fit(x,y,epochs=500)
```

```python
model.get_weights()
```

출력 결과에서 3개의 값을 반환해주는 것을 확인할 수 있는데 3개중 2개는 가중치에 해당하고 나머지 한개는 편향에 해당한다.
